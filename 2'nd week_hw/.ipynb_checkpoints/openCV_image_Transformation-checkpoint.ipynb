```python
import cv2
import numpy as np
import math as m

img = cv2.imread('opencv/picture/newyork.jpg')
## 이미지 화소를 변경했을때, 빈 cell이 생기지 않게 하기 위해서 bilinear interporlation 적용(default값이 항상 bi-linear)
#img = cv2.resize(img, (640,480))
img = cv2.resize(img, (640,480), fx = 1/2, fy = 1/2, interpolation = cv2.INTER_LINEAR)

height, width = img.shape[0], img.shape[1]                                     ## 이미지의 행의 개수(height), 열의 개수(width)

th = m.radians(45)                                                             ## theta 45도 값을 radian으로 설정
m1 = np.array([[1, 0, width/2], [0, 1, height/2], [0,0,1]])                    ## Translation (오른쪽 하단 대각선 방향으로 이동)
m2 = np.array([[m.cos(th), -m.sin(th), 0], [m.sin(th), m.cos(th), 0], [0,0,1]])## Rotation (이미지의 왼쪽 상단 point를 기준으로 회전)
m3 = np.array([[1/2,0,0], [0,1/2,0], [0,0,1]])                                 ## Scaling (이미지 2배 축소)
m4 = np.array([[1, 0, -width/2], [0, 1, -height/2], [0,0,1]])                  ## Translation (왼쪽 상단 대각선 방향으로 이동)

## Scaling-Translation(왼쪽상단 대각선)-Rotation-Translation(오른쪽하단 대각선) 순서로 Transform matrix 생성
M = np.dot(m1, np.dot(m2,np.dot(m3,m4)))
M = M[0:2]

## 이미지 Transform
result = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))

cv2.imshow("img", img)
cv2.imshow("result", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import math as m

img = cv2.imread('opencv/picture/newyork.jpg')
## 이미지 화소를 변경했을때, 빈 cell이 생기지 않게 하기 위해서 bilinear interporlation 적용(default값이 항상 bi-linear)
img = cv2.resize(img, (640,480))
height, width = img.shape[0], img.shape[1]

m = np.array([[1/2,0,0], [0,1/2,0], [0,0,1]])                                 ## Scaling (이미지 2배 축소)
m_inv = np.linalg.inv(m)

img_r = np.zeros((height,width,3), np.uint8)

for h in range(0, img_r.shape[0]):
    for w in range(0, img_r.shape[1]):
        if int(np.dot(m_inv, np.array([h,w,1]))[0]) >= img.shape[0] :
            break
        elif int(np.dot(m_inv, np.array([h,w,1]))[1]) >= img.shape[1] :
            break
        img_r[h,w] = img[int(np.dot(m_inv, np.array([h,w,1]))[0]) , int(np.dot(m_inv, np.array([h,w,1]))[1])]

cv2.imshow("img", img)
cv2.imshow("result", img_r)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import math as m

def bilinear_interpolation(img, x, y):
    a = np.array([0,0,0], np.uint8)
    h,w = int(x), int(y)
    
    X1 = np.dot(1-y+w,img[h,w]) + np.dot(y-w,img[h,w+1])
    X2 = np.dot(1-y+w,img[h+1,w]) + np.dot(y-w,img[h+1,w+1])
    X = np.dot(1-x+h,X1) + np.dot(x-h,X2)

    return X

def Transform(img, trans_M) :
    
    height, width = img.shape[0], img.shape[1]
    img_r = np.zeros((height,width,3), np.uint8)

    for h in range(0, img_r.shape[0]):
        for w in range(0, img_r.shape[1]):
            if int(np.dot(trans_M, np.array([h,w,1]))[0]) >= img.shape[0] :
                break
            elif int(np.dot(trans_M, np.array([h,w,1]))[1]) >= img.shape[1] :
                break
            x, y = np.dot(m_inv, np.array([h,w,1]))[0:2]
            if x-int(x) == 0 and y-int(y):
                img_r[h,w] = img[int(x) , int(y)]
            else :
                img_r[h,w] = bilinear_interpolation(img, x, y)
    
    return img_r

    
img = cv2.imread('../picture/newyork.jpg')
img = cv2.resize(img, (640,480))

m = np.array([[2,0,0], [0,2,0], [0,0,1]])                            ## Scaling (이미지 2배 축소)
m_inv = np.linalg.inv(m)

img_r = Transform(img, m_inv)

cv2.imshow("img", img)
cv2.imshow("result", img_r)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
