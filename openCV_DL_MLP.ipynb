```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OneHotEncoder            ## one hot encoding
import matplotlib.pyplot as plt
import numpy as np
import time
import
mnist = fetch_openml('mnist_784')
encoder = OneHotEncoder()
encoder.fit(mnist.target.reshape(-1,1))
label = encoder.transform(mnist.target.reshape(-1,1)).toarray()

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_deriv(x):
    return sigmoid(x) * (1-sigmoid(x))

def relu(x):
    return np.maximum(x, 0)

def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))

def forward(x,w1,w2,predict=False):
    ## hidden layer와 output layer의 unit은 weight와의 곱으로 생성한다.
    ## 각각의 유닛끼리 한번씩 전부 연결해주므로 매트릭스 연산을한다.
    global h_unit
    h_unit = np.matmul(x,w1)
    h_unit_sig = sigmoid(h_unit)
    #h_unit_relu= relu(h_unit)
    
    o_unit = np.matmul(h_unit_sig, w2)
    o_unit_sig = sigmoid(o_unit)
    #o_unit_relu = relu(o_unit)
    if predict:
        return o_unit_sig
    
    return h_unit, h_unit_sig, o_unit, o_unit_sig

def backprop(o_unit, x, h_unit_sig, o_unit_sig, y):
    ##output과 hidden layer의 Error인 delta와 업데이트할 weight값을 구한다.
    delta2 = o_unit_sig - y                                                        ## 1x10
    Delta2 = np.matmul(h_unit_sig.reshape(-1,1), delta2.reshape(1,-1))             ## (256x1) (1x10) = 256x10
    delta1 = delta2.dot(w2.T) * sigmoid_deriv(h_unit)                              ## (1x10) (10x256) * (1x256)= 1x256
    Delta1 = np.matmul(x.reshape(-1,1), delta1.reshape(1,-1))                      ## (784x1) (1x256) = 784x256
    
    return delta2, Delta1, Delta2


global w2
w1 = np.random.randn(784, 256)
w2 = np.random.randn(256, 10)


########### Learning Start ###############
start = time.time()
for i in range(len(mnist.data[0:1000])):
    train_x = mnist.data[i] / 255
    train_y = label[i]
    
    costs = []
    learning_rate = 0.001
    print("%d번째" %(i+1))
    for i in range(10000):
        h_unit, h_unit_sig, o_unit, o_unit_sig = forward(train_x, w1, w2)
        delta2, Delta1, Delta2 = backprop(o_unit, train_x, h_unit_sig, o_unit_sig, train_y)
        
        w1 -= learning_rate*Delta1
        w2 -= learning_rate*Delta2
        
        c = np.mean(np.abs(delta2))
        costs.append(c)
        
        if i % 1000 == 0 :
            print(f"Iteration: {i}. Error: {c}")
        

end = time.time()        
########### Learning End #################

print("Training complete.")
print("Learning Time : %lf sec" %(end-start))

plt.plot(costs)
plt.show()
############# Test #######################
test_x = mnist.data[1000:1100]
test_y = mnist.target[1000:1100]
predict = []
correct = 0
for x in test_x:
    predict.append(forward(x, w1, w2, True))

for i in range(len(predict)) :
    p = predict[i].tolist()
    if int(test_y[i]) ==  p.index(max(p)):
        correct += 1

accuracy = (correct / len(predict)) * 100
print("accuracy : %lf %%" % accuracy)
print("predict: %s" % predict[0])

cv2.imshow("target image", mnist.data[1000].reshape(28,28))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from sklearn.datasets import fetch_openml
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
import cv2

mnist = fetch_openml('mnist_784')
x_train = mnist.data[0:60000] 
y_train = mnist.target[0:60000] 
x_test = mnist.data[60000:]
y_test = mnist.target[60000:]

mlp = MLPClassifier(solver='adam', activation='relu', random_state=0,learning_rate_init= 0.001, hidden_layer_sizes=[10,10])
mlp.fit(x_train, y_train)
predict = mlp.predict(x_test)

correct = 0
for i in range(len(y_test)):
    if y_test[i] == predict[i] :
        correct += 1

accuracy = (correct / len(predict)) * 100
print("accuracy: %lf %%" % accuracy)
print("predict: %s" % predict[0])

cv2.imshow("target image", mnist.data[60000].reshape(28,28))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
